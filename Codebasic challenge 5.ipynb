### CAGR Calculation Of Domestic Visits

```python
dv_cagr = dv.groupby(['district', 'year'])['visitors'].sum().reset_index()
```

```python
dv_cagr
```

```python
# Function to calculate Cgpr for domestic visitors
def calculate_cagr_for_all_districts(df):
    # Get unique district names from the DataFrame
    district_names = df['district'].unique()

    cagr_values = []  # To store CAGR values for each district

    for district_name in district_names:
        # Filter data for the specific district
        district_data = df[df['district'] == district_name]

        # Check if there is data for the given district
        if district_data.empty:
            print(f"No data available for {district_name}")
            cagr_values.append(np.nan)  # Append NaN for districts with no data
        else:
            # Calculate CAGR
            initial_value = district_data['visitors'].iloc[0]
            final_value = district_data['visitors'].iloc[-1]
            num_years = district_data['year'].iloc[-1] - district_data['year'].iloc[0] 

            cagr = (final_value / initial_value) ** (1 / num_years) - 1

            print(f"The CAGR for {district_name} is: {cagr:.2%}")
            cagr_values.append(cagr)

    # Add a new column 'cagr' to the DataFrame
    df['cagr'] = np.nan  # Initialize the column with NaN
    for district_name, cagr_value in zip(district_names, cagr_values):
        df.loc[df['district'] == district_name, 'cagr'] = cagr_value

# Example usage
# Assuming your data is in a DataFrame named df
calculate_cagr_for_all_districts(dv_cagr)
```

### Q5) Show the top and bottom 3 districts with high domestic to foreign tourist ratio.

```python
def calculate_visitor_ratio(dv, fv):
    """
    Calculate district-wise visitor ratio for each year.

    Parameters:
    - dv: DataFrame with columns 'district', 'year', 'visitors'
    - fv: DataFrame with columns 'district', 'year', 'visitors'

    Returns:
    - DataFrame with columns 'district', 'year', 'visitors_dv', 'visitors_fv', 'ratio'
    """
    # Group by district and year, and sum up visitors
    dv_ratio = dv.groupby(['district'])['visitors'].sum().reset_index()
    fv_ratio = fv.groupby(['district'])['visitors'].sum().reset_index()

    # Merge the two DataFrames on 'district' and 'year'
    df_merged = pd.merge(dv_ratio, fv_ratio, on='district', how='inner', suffixes=('_dv', '_fv'))

    # Calculate the visitor ratio
    df_merged['ratio'] = round(df_merged['visitors_dv'] / df_merged['visitors_fv'],2)

    return df_merged
```

```python
def calculate_future_revenue(initial_visitors, cagr, target_year, average_spending, current_year=2019):
    """
    Calculate estimated revenue for a future year based on CAGR and average spending.

    Parameters:
    - initial_visitors: Initial number of visitors
    - cagr: Compound Annual Growth Rate
    - target_year: Year for which revenue is to be estimated
    - average_spending: Average spending per visitor
    - current_year: Current year (default is 2019)

    Returns:
    - Estimated revenue for the target year
    """
    num_years_future = target_year - current_year
    future_visitors = initial_visitors * (1 + cagr) ** num_years_future
    future_revenue = future_visitors * average_spending
    return int(future_revenue)

# Given data
initial_value = 13802362
cagr = -0.161290  # Use the calculated CAGR
target_year = 2025
average_spending = 1200

# Calculate and print the result
estimated_revenue_2025 = calculate_future_revenue(initial_value, cagr, target_year, average_spending)
print(f"The estimated revenue in {target_year} based on the CAGR and average spending is: {estimated_revenue_2025} Rs")
```
